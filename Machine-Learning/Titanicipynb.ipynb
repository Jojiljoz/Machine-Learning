```python
import numpy as np
import pandas as pd
df = pd.read_csv("/content/Titanic-Dataset.csv")
print(df)
print(df.size)
print(df.shape)
print(df.isna().sum())
print(df.dtypes)
```

```python
df.drop(["Cabin"],axis = 1,inplace = True)
print(df)
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
print(df)
```

```python
print(df.isna().sum())
```

```python
df['Embarked'] = df['Embarked'].fillna(df["Embarked"].mode()[0])
```

```python
print(df.isna().sum())
```

```python
print(df.dtypes)
```

```python
df.drop(["PassengerId","Name"],axis = 1, inplace = True)
```

```python
print(df.dtypes)
```

```python
#To change sec embarked into numerical

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["Sex"] = le.fit_transform(df["Sex"])
df["Embarked"] = le.fit_transform(df["Embarked"])
df
```

```python
print(df.dtypes)
```

```python
print(df)
```

```python
df["Age"] = df["Age"].astype(int)
df["Fare"]  = df["Fare"].astype(int)
print(df)
```

```python
print(df.dtypes)
```

```python
#CHECK OUTLIERS

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize = (8,2))
for col in df.columns:
  sns.boxplot(df[col])
  plt.show()

```

```python
#OUTLIERS OF AGE

Q1 = df["Age"].quantile(0.25)
Q3 = df["Age"].quantile(0.75)

IQR = Q3 - Q1

max_limit = Q3 + (1.5*IQR)
min_limit = Q1 - (1.5*IQR)

df = df[(df["Age"] >= min_limit) & (df["Age"] <= max_limit)]

#OUTLIERS OF SIBSP

Q1 = df["SibSp"].quantile(0.25)
Q3 = df["SibSp"].quantile(0.75)

IQR = Q3 - Q1

max_limit = Q3 + (1.5*IQR)
min_limit = Q1 - (1.5*IQR)

df = df[(df["SibSp"] >= min_limit) & (df["SibSp"] <= max_limit)]

#OUTLIERS OF PARCH

Q1 = df["Parch"].quantile(0.25)
Q2 = df["Parch"].quantile(0.75)

IQR = Q3 - Q1

max_limit = Q3 + (1.5*IQR)
min_limit = Q1 - (1.5*IQR)

df = df[(df["Parch"] >= min_limit) & (df["Parch"] <= max_limit)]

#OUTLIERS OF FARE

Q1 = df["Fare"].quantile(0.25)
Q3 = df["Fare"].quantile(0.75)

IQR = Q3 - Q1

max_limit = Q3 + (1.5*IQR)
min_limit = Q1 - (1.5*IQR)

df = df[(df["Fare"] >= min_limit) & (df["Fare"] <= max_limit)]
```

```python
plt.figure(figsize=(8,2))
for i in df.columns:
  sns.boxplot(df[i])
  plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.30)
x_train
```

```python
x_test
y_train
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
x_train
```

```python
x_test
```

```python
#SVM

from sklearn.svm import SVC
model = SVC()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
y_pred
```

```python
#CLASSIFICATION REPORT

from sklearn.metrics import classification_report
cl = classification_report(y_test,y_pred)
print(cl)
```
